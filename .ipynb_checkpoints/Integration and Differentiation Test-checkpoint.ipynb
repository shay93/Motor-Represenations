```python
from __future__ import division 
import numpy as np
import scipy as sc
import scipy.integrate as integrate
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
time_array = np.linspace(0,5,num = 1025)
noise = np.random.rand(len(time_array)) / 1000
#construct array that represents a polynomial 
a = 6
b = 2
y = a*(time_array ** 2) + b*time_array + noise
y_integrated = (a/3)*(time_array ** 3) + (b/2)*(time_array ** 2)

plt.figure()
plt.plot(time_array,y,label = "y")
plt.plot(time_array,y_integrated,label = "integral")
plt.plot()
plt.legend()
plt.show()
```

```python
#now check whether it is possible to integrate or differentiate the above
#initialize an ode object
dt = 5 /1025
y0 = [0]
t0 = 0
index = 0
def int_func(t,y):
    return a*(t**2) + b*t

r = integrate.ode(int_func).set_integrator("lscope")
r.set_initial_value(0,0)
i = 0
calc_y = np.zeros(1025)
while r.t < 5:
    temp = r.integrate(r.t + dt)
    calc_y[i] = temp
    i+=1
    
print len(np.diff(calc_y))
print len(calc_y)
    
plt.figure()
plt.plot(time_array,y,label= "y")
plt.plot(time_array,calc_y,label = "integrated")
plt.plot(time_array,np.concatenate(([0],np.diff(calc_y) / dt),label = "derivative")
plt.legend()
plt.show()
torque_index = 0
```
