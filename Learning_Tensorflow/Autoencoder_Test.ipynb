```python
from __future__ import division
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

% matplotlib inline
```

```python
IMG_WIDTH = 64
DIRECTORY_NAME = 'Training_Images/'
BATCH_SIZE = 1
EPOCHS = 1
PIXEL_DEPTH = 255
```

```python
def extract_data():
    #initialize numpy array to hold batch of images
    data_batch = np.zeros([BATCH_SIZE,IMG_WIDTH,IMG_WIDTH,1])
    for i in range(BATCH_SIZE):
        shape_ind = np.round(np.random.rand()*2)
        if shape_ind == 0:
            shape_str = 'Triangle'
        elif shape_ind == 1:
            shape_str = 'Rectangle'
        else:
            shape_str = 'Square'

        index = int(np.round(np.random.rand()*999)) 
        #read the input image and place it in in corresponding position of data_batch array
        data_batch[i,:,:,0] = (plt.imread(DIRECTORY_NAME + shape_str + str(index) + '.png'))/255
    return data_batch
```

```python
def build_graph():
    #define placeholder for input data
    x = tf.placeholder('float',shape = [BATCH_SIZE,IMG_WIDTH,IMG_WIDTH,1])
    #define parameters for first convolutional layer
    W_conv1 = tf.Variable(tf.truncated_normal([6,6,1,8],stddev = 0.1))
    b_conv1 = tf.Variable(tf.constant(0.1,shape = [8]))
    #define parameters for the second convolutional layer
    W_conv2 = tf.Variable(tf.truncated_normal([6,6,8,4],stddev = 0.1))
    b_conv2 = tf.Variable(tf.constant(0.1,shape = [4]))
    #define parameters for full connected layer
    W_fc = tf.Variable(tf.truncated_normal(shape = [IMG_WIDTH*IMG_WIDTH*4*BATCH_SIZE,BATCH_SIZE*IMG_WIDTH*IMG_WIDTH],stddev = 0.1))
    b_fc = tf.Variable(tf.constant(0.1,shape = [BATCH_SIZE*IMG_WIDTH*IMG_WIDTH]))
    #now build the operations for the graph
    h_conv1 = tf.nn.tanh(tf.nn.conv2d(x,W_conv1,strides = [1,1,1,1],padding = 'SAME'))
    h_conv2 = tf.nn.tanh(tf.nn.conv2d(h_conv1,W_conv2,strides = [1,1,1,1],padding = 'SAME'))
    h_conv2_flat = tf.reshape(h_conv2, shape = [1,IMG_WIDTH*IMG_WIDTH*BATCH_SIZE*4])
    y_reshape = tf.nn.tanh(tf.matmul(h_conv2_flat,W_fc) + b_fc)
    y = tf.reshape(y_reshape, shape = [BATCH_SIZE,IMG_WIDTH,IMG_WIDTH,1])
    #now define output place holder and loss function
    y_ = tf.placeholder('float', shape = [BATCH_SIZE,IMG_WIDTH,IMG_WIDTH,1])
    #define loss function for training
    meansq = tf.reduce_mean(tf.square(y_-y))
    #add a summary statistic operation
    msq_sum = tf.scalar_summary('meansq',meansq)
    train_step = tf.train.AdamOptimizer(1e-4).minimize(meansq)
    return y,x,y_,meansq,train_step,W_conv1,W_conv2,b_conv1,b_conv2,W_fc,b_fc,h_conv1,h_conv2
```

```python
def run():
    loss = 0
    y,x,y_,meansq,train_step,W_conv1,W_conv2,b_conv1,b_conv2,W_fc,b_fc,h_conv1,h_conv2 = build_graph()
    sess = tf.Session()
    sess.run(tf.initialize_all_variables())
    data_batch = extract_data()
    num_steps = (EPOCHS * 3 * 1000) // BATCH_SIZE  
    
    for step in range(num_steps):
        data_batch = extract_data()
        sess.run(train_step, feed_dict = {x : data_batch, y_ : data_batch})
        loss += sess.run(meansq, feed_dict = {x : data_batch, y_ : data_batch})
        if step % 100 == 0:
            print step,loss
            loss = 0

    
    return sess,y,x,y_,meansq,train_step,W_conv1,W_conv2,b_conv1,b_conv2,W_fc,b_fc,h_conv1,h_conv2
```

```python
sess,y,x,y_,meansq,train_step,W_conv1,W_conv2,b_conv1,b_conv2,W_fc,b_fc,h_conv1,h_conv2y, y_= run()
```
